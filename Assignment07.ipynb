```python
# Extend on O-level or build your own :)
#OOP method
import string

# functions 
class TextAnalysis:
    def __init__(self, link):
      self.link = link          #link to text file
      infile = open(link, 'r')  #open text file, copy to lines
      self.lines = infile.readlines() #read text into a list of lines
      infile.close()
      #list of stopwords
      self.stopwords = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]  
      self.wordlist = {} #empty dictionary, to store unique words as keys and word count as values

    def menu(self):             #menu for choice
        print('EXCITING TEXT ANALYSIS')
        print('1. Display text')
        print("2. Remove stop words")
        print('3. Remove unique words')
        print("4. Convert to lowercase")  
        print('5. Word Count')
        print("6. Find most frequently occuring term")
        print('7. Find N most frequently occuring terms')
        print('8. Find # of occurence of specified word')       
        print("0. Exit")

    def display(self):          #display lines, with whatever modifications made
        for line in self.lines: #calling each item in the list of lines
            print(line)
        print()

    def remove_stop_words(self):  #remove all stop words from lines
        self.remove_words(self.stopwords)
        print('Stop words removed.')
        self.wordcount() #update wordcount
        print()

    def remove_unique_words(self, numoftimes):  #remove unique words
        newlist = [] #new list to store unique words
        for k, v in self.wordlist.items():  #filter unique words
            if v <= numoftimes:
                newlist.append(k)  
        self.remove_words(newlist)
        print('Unique words removed.')
        self.wordcount() #update wordcount
        print()

    def remove_words(self, A): #remove words in list A
        for i in range(len(self.lines)):
            #removing all punctuation, convert to lowercase, and then split string to list
            lst = self.lines[i].translate(str.maketrans('', '', string.punctuation)).lower().split()            
            for word in A:
                lst = [value for value in lst if value != word]   #removing words from lst 
            self.lines[i] = ' '.join(lst)                         #joining list to form a single string, then overwrite existing line in lines

    def convert_lower_case(self): #covert all words to lower case
        for i in range(len(self.lines)):              
            self.lines[i] = self.lines[i].lower()
        print('Converted to lower case.')
        print()

    def wordcount(self):  #storing count of unique words in dictionary, to be accessed when needed
        self.wordlist = {} #reset to zero
        for line in self.lines:
            #removing all punctuation, convert to lowercase, and then split string to list
            lst = line.translate(str.maketrans('', '', string.punctuation)).lower().split()
            for word in lst:
                if word in self.wordlist.keys():
                    self.wordlist[word] += 1
                else:
                    self.wordlist[word] = 1

    def most_occuring_term(self): #search for most occuring term in lines
        #self.wordcount()
        mostterm = max(self.wordlist, key = self.wordlist.get)  #search for key with largest value - mostterm is the most occuring term
        print(f"The most occuring term is '{mostterm}', and it appeared {self.wordlist[mostterm]} times.")
        print()
    
    def n_most_terms(self, num):
        newcount = sorted(self.wordlist.items(), key = lambda x:-x[1])[:num]

        print(f'The {num} most occurring terms with word count:')
        for word in newcount:
            print(f'{word[0]:12}: {word[1]}')
        print()
    
    def totalwordcount(self):  #count number of words in lines
        #self.wordcount()
        print(f'Word count: {sum(self.wordlist.values())}')
        print()

    def wordsearch(self, text): #count number of occurence of text in lines       
        #self.wordcount()
        if text in self.wordlist.keys():
            print(f"Word count of '{text}': {self.wordlist[text]}")
        else:
            print('Word not found in text.')
        print()

    def initialise(self): #run TA
        self.wordcount()
        option = ''
        while option != '0':
          self.menu()
          option = input("Enter option: ")
          if option == '1':
              self.display()
          elif option == '2':
              self.remove_stop_words()
          elif option == '3':
              n = int(input('Enter number of occurence of each unique word (the smaller, the more unique the word): '))
              self.remove_unique_words(n)    
          elif option == '4':
              self.convert_lower_case()
          elif option == '5':
              self.totalwordcount()
          elif option == '6':
              self.most_occuring_term()
          elif option == '7':
              n = int(input('Enter N (in integers): '))
              self.n_most_terms(n)
          elif option == '8':
              word = input('Please enter word: ')
              self.wordsearch(word)

          
          
          elif option == '0': # quit
              print("Thank you for using EXCITING TEXT ANALYSIS!")
          else: # invalid
              print("Please enter a valid option.")
              print()
        print("Bye.")

# main
ta = TextAnalysis('/content/drive/MyDrive/Colab Notebooks/cos22.txt')
ta.initialise()
```
