```python
# import csv and random libraries
import csv
import random

def menu():
  #main menu options
  print("Super Search Engine")
  print('1. Display first few rows of data')
  print('2. Search for first occurance on exact match')
  print('3. Search for all occurance on exact match')
  print('4. Search for all occurance on partial match')
  print('5. Magic search - Random search for occurance on exact match')
  print('6. Count distinct values in specific columns')
  print('0. Exit')
  print()

def show_nrows(n):
    #display top n lines
    counter = 0         #counter for counting number of rows printed
    with open("/content/drive/MyDrive/Colab Notebooks/primary-enrolment-by-age.csv", 'r', newline="") as infile:
        records = csv.reader(infile, delimiter=",")
        for row in records:
            print (row)
            counter += 1        #counting the number of times row was printed
            if counter == n+1:  #breaks at n rows + 1 row of column headers
                break
    print()

def search_one(target):
    #search for first occurrence
    with open("/content/drive/MyDrive/Colab Notebooks/primary-enrolment-by-age.csv", 'r', newline="") as infile:
        records = csv.reader(infile, delimiter=",")
        for row in records:
            if target in row:   #search if target matches each item of row list
                print (row)
                break           #breaks after first occurance
    print()
    

def search_many(target):
    #search for all occurrences
    with open("/content/drive/MyDrive/Colab Notebooks/primary-enrolment-by-age.csv", 'r', newline="") as infile:
        records = csv.reader(infile, delimiter=",")
        for row in records:
            if target in row:   #search if target matches each item of row list
                print (row)     #prints only if true
    print()

def search_partial(target):
    #search partial match
    with open("/content/drive/MyDrive/Colab Notebooks/primary-enrolment-by-age.csv", 'r', newline="") as infile:
        records = csv.reader(infile, delimiter=",")
        for row in records:
            for items in row:         #nested loop to look through each item in each row list
                if target in items:   #search if target is a substring of string in items
                    print (row)
    print()

def search_magic(target):
    #search for magic - random search
    with open("/content/drive/MyDrive/Colab Notebooks/primary-enrolment-by-age.csv", 'r', newline="") as infile:
        records = csv.reader(infile, delimiter=",")
        lst = []                                  #empty list to copy records over
        for row in records:
            lst.append(row)                       #appending each row to list
        a = random.randint(1, len(lst)-1)         #selecting random row number
        b = random.randint(1, len(lst[0])-1)      #selecting random column number
        while not target in lst[a][b]:
            a = random.randint(1, len(lst)-1)     #row again if target does not match any item in row list
            b = random.randint(1, len(lst[0])-1) 
        print (lst[a])
    print()

def count_in_columns():
    #display distinct data in specified coloumns
    with open("/content/drive/MyDrive/Colab Notebooks/primary-enrolment-by-age.csv", 'r', newline="") as infile:
        records = csv.reader(infile, delimiter=",")
        lst = []                                  #empty list to copy records 
        for row in records:
            lst.append(row)                       #append each row to list
        for i, v in enumerate(lst[0]):            #enumerate each item in first row, containing the column headers
            print(str(i+1)+ '. '+lst[0][i])
        col = -1                                  #defining col for user to input column number
        while not 0<col<=len(lst[0]):             #loops until valid column number chosen
            col = int(input('Please key in column number to display: '))
        display={}                                #empty dictionary, where column entry as key and item count as value
        for row in lst:
            if row[col-1] in display.keys():
                display[row[col-1]] += 1          #add count by one if key already exists in dict
            else:
                display[row[col-1]] = 1           #add new key with value 1
        print(display)
    print()

# main
option = ''
while option != '0': 
  # display menu
  menu()
  # get option
  option = input("Enter option:")
  if option == '1':
    num = int(input("Enter number of rows: "))
    show_nrows (num)
  elif option == '2':
    query = input("Enter query: ")
    search_one(query)
  elif option == '3':
    query = input("Enter query: ")
    search_many(query)
  elif option == '4':
    query = input("Enter query: ")
    search_partial(query)
  elif option == '5':
    query = input("Enter query: ")
    search_magic(query)
  elif option == '6':
   count_in_columns()
  elif option == '0':
    print("Bye")
  else: # invalid option
    print("Not implemented yet? Please choose a valid option.")
print("Thank you for using Super Search Engine!")
```
